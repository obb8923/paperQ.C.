```python
#일반화 결과

%matplotlib inline
import numpy as np
from qiskit import QuantumCircuit, transpile
from qiskit.visualization import plot_histogram
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
from qiskit.circuit.library import UnitaryGate
import matplotlib.pyplot as plt

w = 1.2
r = 3

t_arr = np.linspace(0,10,1000)
arr_00_G=[]
arr_01_G=[]
arr_10_G=[]
arr_11_G=[]


for i in range(np.size(t_arr)):
    matrix_jc= np.array([[np.exp(1j*w*t_arr[i]),0,0,0],
                  [0,np.cos(r*t_arr[i]),1j*np.sin(r*t_arr[i]),0],
                  [0,1j*np.sin(r*t_arr[i]),np.cos(r*t_arr[i]),0],
                  [0,0,0,np.exp(-1j*w*t_arr[i])]])
    
    qc = QuantumCircuit(2,2)
    #qc.x(0)
    qc.x(1)
    qc.barrier()
    
    gate=UnitaryGate(matrix_jc)
    qc.append(gate,[0,1])    
    ket = Statevector(qc)
    arr_00_G.append(np.abs(ket[0])**2)
    arr_01_G.append(np.abs(ket[1])**2)
    arr_10_G.append(np.abs(ket[2])**2)
    arr_11_G.append(np.abs(ket[3])**2)


plt.plot(t_arr, arr_00_G, label= '00')
plt.plot(t_arr, arr_01_G, label= '01')
plt.plot(t_arr, arr_10_G, label= '10')
plt.plot(t_arr, arr_11_G, label= '11')
plt.title('Generalization')
plt.xlabel('Time')
plt.ylabel('Probaility')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#근사 디컴 결과

import numpy as np
from qiskit import QuantumCircuit, transpile
from qiskit.quantum_info import Operator, Statevector
import matplotlib.pyplot as plt

# Define the parameters
w = 1.2
r = 3
dt = 0.001

# Define the matrix
matrix = np.array([[1 + 1j * w * dt - 0.5 * (dt ** 2) * (w ** 2), 0, 0, 0],
                   [0, 1 - 0.5 * (dt ** 2) * (r ** 2), 1j * r * dt, 0],
                   [0, 1j * r * dt, 1 - 0.5 * (dt ** 2) * (r ** 2), 0],
                   [0, 0, 0, 1 - 1j * w * dt - 0.5 * (dt ** 2) * (w ** 2)]])

# Convert the matrix to Operator
op = Operator(matrix)

# Define the time array and result arrays
t_arr = np.linspace(0, 10000, 1000)
arr_00 = []
arr_01 = []
arr_10 = []
arr_11 = []

# Simulate the quantum circuit for different time steps
for i in range(len(t_arr)):
    qc = QuantumCircuit(2, 2)
    qc.x(1)
    qc.barrier()
    
    
    for _ in range(int(t_arr[i])):
        qc.unitary(op, [0, 1])
    
    transpiled_circuit = transpile(qc, basis_gates=['ry', 'rz', 'cx'])
    
    ket = Statevector.from_instruction(transpiled_circuit)
    arr_00.append(np.abs(ket[0])**2)
    arr_01.append(np.abs(ket[1])**2)
    arr_10.append(np.abs(ket[2])**2)
    arr_11.append(np.abs(ket[3])**2)

# Plot the results
plt.plot(t_arr, arr_00, label='00')
plt.plot(t_arr, arr_01, label='01')
plt.plot(t_arr, arr_10, label='10')
plt.plot(t_arr, arr_11, label='11')
plt.title('Approximation Decomposition')
plt.xlabel('Time')
plt.ylabel('Probability')
plt.legend()
plt.grid(True)
plt.show()
```
