```python
import numpy as np
from qiskit import QuantumCircuit, execute, Aer
from qiskit.visualization import plot_histogram
from qiskit.extensions import *
from qiskit.quantum_info import Statevector

#  
w = 26
r = 2

# t   
t_values = np.linspace(0, 2*np.pi, 100)  # : t 0 2*pi 100  

# Unitary gate  
def get_matrix(t):
    return np.array([[np.exp(1j*w*t), 0, 0, 0],
                     [0, np.cos(r*t), 1j*np.sin(r*t), 0],
                     [0, 1j*np.sin(r*t), np.cos(r*t), 0],
                     [0, 0, 0, np.exp(-1j*w*t)]])

#  Statevector   
statevectors = []

#  t   
for t in t_values:
    matrix = get_matrix(t)
    gate = UnitaryGate(matrix)
    
    #   
    circuit = QuantumCircuit(2, 2)
    circuit.append(gate, [0, 1])
    
    #   Statevector 
    ket = Statevector(circuit)
    
    # Statevector  
    statevectors.append(ket)

#  'statevectors'   t   Statevector .
#  Statevector        .
import matplotlib.pyplot as plt

# statevectors  statevector     
for ket in statevectors:
    probs = np.abs(ket.data)**2
    plt.bar(range(len(probs)), probs)
    plt.xlabel('')
    plt.ylabel('')
    plt.title('Statevector ')
    plt.show()
```

```python
%matplotlib inline
import numpy as np
from qiskit import QuantumCircuit, execute, Aer
from qiskit.visualization import plot_histogram
from qiskit.extensions import *
from qiskit.quantum_info import Statevector
import matplotlib.pyplot as plt

w = 1
r = 1

t_arr = np.linspace(0,10,100)
arr_00=[]
arr_01=[]
arr_10=[]
arr_11=[]


for i in range(np.size(t_arr)):
    matrix_jc= np.array([[np.exp(1j*w*t_arr[i]),0,0,0],
                  [0,np.cos(r*t_arr[i]),1j*np.sin(r*t_arr[i]),0],
                  [0,1j*np.sin(r*t_arr[i]),np.cos(r*t_arr[i]),0],
                  [0,0,0,np.exp(-1j*w*t_arr[i])]])
    
    qc = QuantumCircuit(2,2)
    #qc.x(0)
    qc.x(1)
    qc.barrier()
    
    gate=UnitaryGate(matrix_jc)
    qc.append(gate,[0,1])    
    ket = Statevector(qc)
    arr_00.append(np.abs(ket[0])**2)
    arr_01.append(np.abs(ket[1])**2)
    arr_10.append(np.abs(ket[2])**2)
    arr_11.append(np.abs(ket[3])**2)


plt.plot(t_arr, arr_00, label= '00')
plt.plot(t_arr, arr_01, label= '01')
plt.plot(t_arr, arr_10, label= '10')
plt.plot(t_arr, arr_11, label= '11')
plt.xlabel('Time')
plt.ylabel('Probaility')
plt.legend()
plt.show()
```

```python
import numpy as np
from qiskit import QuantumCircuit, transpile
from qiskit.quantum_info import Operator

# Define the parameters
w = 1.2
r = 3
dt = 0.001

# Define the matrix
matrix = np.array([[1+1j*w*dt-0.5*(dt**2)*(w**2), 0, 0, 0],
                   [0, 1-0.5*(dt**2)*(r**2), 1j*r*dt, 0],
                   [0, 1j*r*dt, 1-0.5*(dt**2)*(r**2), 0],
                   [0, 0, 0, 1-1j*w*dt-0.5*(dt**2)*(w**2)]])

# Convert the matrix to Operator
op = Operator(matrix)

# Create a QuantumCircuit
qc = QuantumCircuit(2)

# Initialize the circuit with the operator
qc.unitary(op, [0, 1])

# Transpile the circuit into the desired basis gates
transpiled_circuit = transpile(qc, basis_gates=['ry', 'rz', 'cx'])

print(transpiled_circuit)
```
